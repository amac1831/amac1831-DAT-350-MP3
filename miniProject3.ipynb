```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# load in the data

rawDF = pd.read_csv("Titanic-Dataset.csv.xls")

rawDF.head()
```

```python
rawDF["Survived"].value_counts()
```

```python
def printAmtNA(aDF):
   for aCol in aDF:
    print(f"The amount of missing values in {aCol} is {aDF[aCol].isna().sum()}")

printAmtNA(rawDF)
```

```python
droppedDF = rawDF.drop(columns=['Cabin'])

droppedDF = droppedDF.drop(columns=['Name'])

droppedDF = droppedDF.drop(columns=['Ticket'])

droppedDF
```

```python
from sklearn.impute import SimpleImputer

import numpy as np

imputer = SimpleImputer(strategy="mean")

imputedDF = droppedDF.copy()
imputedDF["Age"] = imputer.fit_transform(imputedDF[["Age"]])

imputedDF.tail()
```

```python
rowRemoveDF = imputedDF.dropna(axis=0)

rowRemoveDF
```

```python
categoricalCols = ["Sex", "Embarked", "Pclass"]

categoricalCols
```

```python
from sklearn.preprocessing import OneHotEncoder

# instantiate and fit the one hot encoder
encoder = OneHotEncoder(sparse_output=False)
encoderArray = encoder.fit_transform(rowRemoveDF[categoricalCols])
colNames = encoder.get_feature_names_out()

# fix the shape
colNames = np.array(colNames).ravel()

# get the dataframe of the 
oneHotDF = pd.DataFrame(encoderArray, columns=colNames)

oneHotDF.head()
```

```python

rowRemoveDF = rowRemoveDF.drop(columns = categoricalCols)

rowRemoveDF = pd.concat([rowRemoveDF, oneHotDF], axis=1)

rowRemoveDF.head()
```

```python
processedDF = rowRemoveDF.dropna(axis=0)

processedDF
```

```python
yDF = processedDF["Survived"]
xDF = processedDF.drop(columns=["Survived"], axis=1)

print(xDF.columns)
xDF.head()
```

```python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(
    xDF, yDF, 
    test_size=0.05,
    shuffle = True,
    stratify = yDF)
```
